```python
import numpy as np 
import pandas as pd 
import re
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.neural_network import MLPClassifier

############### Loading data #####################
TrainingData = pd.read_csv('train.csv')
Final_test_set = pd.read_csv('test.csv')
############## Preprocessing data ####################
############## Preprocessing data ####################
X = TrainingData
#Removing columns of the training data which we have not yet found a way to convert into meaningful numerical expressions. 
X = X.drop('Name',1)
X = X.drop('Ticket',1)
```

```python
#replacing male = 1 and female = 0.
X = X.replace(to_replace = 'male',value= 1)
X = X.replace(to_replace = 'female',value = 0) 

# replace nan in age by the average age on the boat
av_age = X['Age'].mean()
X['Age'] = X['Age'].replace(to_replace =np.NaN,value=av_age) // 1

# replace the list of embarked 
# list(set(X['Embarked']))
#[nan, 'S', 'C', 'Q']
X['Embarked'] = X['Embarked'].replace(to_replace =np.NaN,value=0.0)
X['Embarked'] = X['Embarked'].replace(to_replace ='S',value=1.0)
X['Embarked'] = X['Embarked'].replace(to_replace ='C',value=2.0)
X['Embarked'] = X['Embarked'].replace(to_replace ='Q',value=3.0)

# replace the list of embarked 
X['Cabin'] = X['Cabin'].replace(to_replace =np.NaN,value=0.0)
Ca = X['Cabin'].tolist();
p1=[]
p2=[]
p1.append(0)
p2.append(0)
for x in range(1,len(X.index)):
    #print("x=",x)
    #print('Ca=',Ca[x])
    if Ca[x] == 0:
        p1.append(0)
        p2.append(0)
    else:
        #print("len=",len(re.findall('[a-zA-Z]+|\\d+', Ca[x])))
        if len(re.findall('[a-zA-Z]+|\\d+', Ca[x])) == 3:
            p1.append(re.findall('[a-zA-Z]+|\\d+', Ca[x])[1])
            p2.append(re.findall('[a-zA-Z]+|\\d+', Ca[x])[2])
        else:
            p1.append(re.findall('[a-zA-Z]+|\\d+', Ca[x])[0])

            if len(re.findall('[a-zA-Z]+|\\d+', Ca[x])) == 1:
                p2.append(0)
            else:
                p2.append(re.findall('[a-zA-Z]+|\\d+', Ca[x])[1])
X['Cabin_Letter']=pd.Series(p1)
X['Cabin_Number']=pd.Series(p2)
#X = X.drop('Cabin',1)
#list(set(X['Cabin_Letter']))
#[0, 'T', 'D', 'E', 'C', 'F', 'B', 'A', 'G'
X['Cabin_Letter'] = X['Cabin_Letter'].replace(to_replace ='A',value=1.0)
X['Cabin_Letter'] = X['Cabin_Letter'].replace(to_replace ='B',value=2.0)
X['Cabin_Letter'] = X['Cabin_Letter'].replace(to_replace ='C',value=3.0)
X['Cabin_Letter'] = X['Cabin_Letter'].replace(to_replace ='D',value=4.0)
X['Cabin_Letter'] = X['Cabin_Letter'].replace(to_replace ='E',value=5.0)
X['Cabin_Letter'] = X['Cabin_Letter'].replace(to_replace ='F',value=6.0)
X['Cabin_Letter'] = X['Cabin_Letter'].replace(to_replace ='G',value=7.0)
X['Cabin_Letter'] = X['Cabin_Letter'].replace(to_replace ='T',value=7.0)
```

```python
#Extracting the y-column with the list of who survived
y = X['Survived']
X = X.drop('Survived',1)
#split the TrainingData into a test and training set
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)

############### Data Analysis #####################
#Decision tree classfier
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
#clf.predict(X_test)
print('Decision Tree Classfier score',clf.score(X_test,y_test))
#Support Vector Machine classfier

clf = svm.SVC()
clf.fit(X_train,y_train)
print('Support Vector Machine classifier score',clf.score(X_test,y_test))
```
